```python
pbp_pass = pbp[pbp['play_type'] == 'pass'].copy()
pbp_pass = pbp_pass.sort_values(['old_game_id', 'play_id'])
```

### Calculated Fields

Using a rolling window of last 120 pass plays because that is about how many dropbacks there are in three games. I tried it with just 40 dropbacks and the rate varied too much from individual sacks.

```python
pbp_pass['rolling_qb_sack_rate'] = (
    pbp_pass
    .groupby('passer_player_id')['sack']
    .transform(lambda x: x.shift(1).rolling(window=120, min_periods=1).mean())
)
```

```python
pbp_pass['rolling_offense_sack_rate'] = (
    pbp_pass
    .groupby('posteam')['sack']
    .transform(lambda x: x.shift(1).rolling(window=120, min_periods=1).mean())
)
```

```python
pbp_pass['rolling_defense_sack_rate'] = (
    pbp_pass
    .groupby('defteam')['sack']
    .transform(lambda x: x.shift(1).rolling(window=120, min_periods=1).mean())
)
```

### Merging Data

```python
pbp_model = pbp_pass[[
    'old_game_id',
    'play_id',
    'home_team',
    'away_team',
    'season',
    'season_type',
    'week',
    'posteam',
    'yardline_100',
    'half_seconds_remaining',
    'game_seconds_remaining',
    'game_half',
    'down',
    'ydstogo',
    'play_type',
    'shotgun',
    'no_huddle',
    'qb_dropback',
    'posteam_timeouts_remaining',
    'defteam_timeouts_remaining',
    'posteam_score',
    'defteam_score',
    'score_differential',
    'fg_prob',
    'td_prob',
    'wp',
    'sack',
    'passer_player_id',
    'passer_player_name',
    'cp',
    'play_clock',
    'roof',
    'wind',
    'rolling_qb_sack_rate',
    'rolling_offense_sack_rate',
    'rolling_defense_sack_rate'
]].copy()
```

```python
roster_model = roster[[
    'gsis_id',
    'full_name',
    'team',
    'season',
    'position',
    'depth_chart_position',
    'years_exp'
]].copy()
```

```python
temp_merge = pbp_model.merge(
    roster_model,
    left_on=['passer_player_id', 'season', 'posteam'],
    right_on=['gsis_id', 'season', 'team'],
    how='left'
)

temp_merge.drop(columns=['gsis_id'], inplace=True)
```

```python
temp_merge = temp_merge.sort_values(
    ['old_game_id', 'play_id']
).reset_index(drop=True)
```
